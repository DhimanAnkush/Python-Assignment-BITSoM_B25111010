```python
# Raw dataset provided in the prompt
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_roster = []

for entry in raw_students:
    # 1. Clean the name: strip whitespace and apply Title Case
    clean_name = entry['name'].strip().title()
    
    # 2. Convert roll to integer
    clean_roll = int(entry['roll'])
    
    # 3. Parse marks string into a list of integers
    clean_marks = [int(m.strip()) for m in entry['marks_str'].split(',')]
    
    # 4. Name Validation (check if words are alphabetic)
    # Using replace(' ', '') to ignore the spaces between first and last names
    is_valid = clean_name.replace(' ', '').isalpha()
    status = "✓ Valid name" if is_valid else "✗ Invalid name"
    
    # Store cleaned data
    student_profile = {
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks
    }
    cleaned_roster.append(student_profile)
    
    # 5. Print the Profile Card
    print(f"{status}")
    print("================================")
    print(f"Student : {student_profile['name']}")
    print(f"Roll No : {student_profile['roll']}")
    print(f"Marks   : {student_profile['marks']}")
    print("================================\n")

# 6. Specific requirement for Roll 103
for s in cleaned_roster:
    if s['roll'] == 103:
        target_name = s['name']
        print(f"Roll 103 (UPPER): {target_name.upper()}")
        print(f"Roll 103 (lower): {target_name.lower()}")
```

```python
# Initial Data
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print(f"--- Grade Report for {student_name} ---")

# 1. Loop through and assign grades
for sub, score in zip(subjects, marks):
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{sub:10}: {score} ({grade})")

# 2. Calculations
total_sum = sum(marks)
avg_score = round(total_sum / len(marks), 2)

# Finding Max/Min using index to get the subject name too
max_val = max(marks)
max_sub = subjects[marks.index(max_val)]

min_val = min(marks)
min_sub = subjects[marks.index(min_val)]

print("-" * 30)
print(f"Total Marks: {total_sum}")
print(f"Average:     {avg_score}")
print(f"Highest:     {max_sub} ({max_val})")
print(f"Lowest:      {min_sub} ({min_val})")
print("-" * 30)

# 3. Interactive While Loop
print("\n--- Add New Subjects (Type 'done' to stop) ---")
new_count = 0

while True:
    new_sub = input("Enter subject name: ").strip()
    if new_sub.lower() == 'done':
        break
    
    score_input = input(f"Enter marks for {new_sub}: ")
    
    # Simple validation to prevent crashes
    try:
        new_mark = float(score_input)
        if 0 <= new_mark <= 100:
            subjects.append(new_sub)
            marks.append(new_mark)
            new_count += 1
            print(f"Added {new_sub} successfully.")
        else:
            print("Warning: Marks must be between 0 and 100. Entry ignored.")
    except ValueError:
        print("Warning: Please enter a valid numeric value for marks.")

# Final Summary
final_avg = round(sum(marks) / len(marks), 2)
print(f"\nUpdate Complete.")
print(f"New subjects added: {new_count}")
print(f"New overall average: {final_avg}")
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Containers for summary stats
processed_stats = []
pass_count = 0
fail_count = 0

# Table Header
print(f"{'Name'.ljust(18)} | {'Average'.center(7)} | Status")
print("-" * 40)

for name, scores in class_data:
    # Calculate average for current student
    avg = round(sum(scores) / len(scores), 2)
    
    # Determine Pass/Fail status
    status = "Pass" if avg >= 60 else "Fail"
    
    # Update counters
    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1
    
    # Store for final summary
    processed_stats.append({"name": name, "avg": avg})
    
    print(f"{name.ljust(18)} |  {avg:05.2f}  | {status}")

# --- Final Summary Calculations ---

# Find Topper
topper = processed_stats[0]
for s in processed_stats:
    if s['avg'] > topper['avg']:
        topper = s

# Class Average
all_averages = [s['avg'] for s in processed_stats]
overall_avg = round(sum(all_averages) / len(all_averages), 2)

print("-" * 40)
print(f"Passed: {pass_count} | Failed: {fail_count}")
print(f"Class Topper: {topper['name']} ({topper['avg']})")
print(f"Overall Class Average: {overall_avg}")
```

```python
# Initial messy string
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. Base Cleaning: Strip whitespace
clean_essay = essay.strip()
print(f"Base Cleaned Essay: '{clean_essay}'")

# 2. Title Case
print(f"Title Case Result: {clean_essay.title()}")

# 3. Count "python" 
py_count = clean_essay.count("python")
print(f"Occurrences of 'python': {py_count}")

# 4. Replace with Emoji
updated_essay = clean_essay.replace("python", "Python 🐍")
print(f"Replaced Version: {updated_essay}")

# 5. Split into sentences
# Splitting by period + space as requested
sentence_list = clean_essay.split(". ")
print(f"Sentence List: {sentence_list}")

# 6. Numbered Formatting
print("\n--- Formatted Sentences ---")
for i, sentence in enumerate(sentence_list, 1):
    # Ensure each sentence is stripped and ends with a period
    final_sentence = sentence.strip()
    
    if not final_sentence.endswith("."):
        final_sentence += "."
        
    print(f"{i}. {final_sentence}")
```
